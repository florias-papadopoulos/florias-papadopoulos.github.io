### Δημιουργία στήλης δραστηριότητας

Σε πρώτη φάση, γίνεται η εύρεση της δραστηριότητας που έχει πλειοψηφία με την συνάρτηση "most_common_type" (για κάθε 1000 σείρες) και έπειτα δημιουργήθηκε μια νέα στήλη "Activity" που περιέχει αυτές τις "δραστηριότητες με πλειοψηφία". </br>

Υ.Γ. Κάθε 50 χρονικές στιγμές αντιστοιχούν σε 1 δευτερόλεπτο, όποτε καθώς τα παράθυρα είναι των 20 δευτερολέπτων, το καθένα τους θα περιέχει 1000 χρονικές στιγμές. Ακόμη, καθώς τα δεδομένα έχουν 63.000 χρονικές στιγμές στο σύνολο τους, θα έχουμε τελικώς 63 παράθυρα (και άρα και 63 πρότυπα ανά Participant, μετά από επόμενα βήματα).

```python
# Define a function to find the most common type in each group of 1000 rows
def most_common_type(df):
    return df['Type'].value_counts().idxmax()

# Group the DataFrame into groups of 1000 rows and find the most common type in each group
common_types = par.groupby(par.index // 1000).apply(most_common_type)

# Create a new DataFrame with the most common type for every 1000 rows
df_common_types = pd.DataFrame({'Type': common_types})

# Reset the index of the new DataFrame to create a new column
df_common_types = df_common_types.reset_index(drop=True)

# Add the new column to the original DataFrame
par['Activity'] = df_common_types['Type']

par = par.drop(par.columns[0], axis=1)

par
```

### Εξαγωγή χαρακτηριστικών

Σε αυτό το σημείο, με παρόμοια διαδικασία για κάθε χαρακτηριστικό, δημιουργήθηκαν οι αντίστοιχες στήλες για το καθένα. Τα χαρακτηριστικά που υπολογίστηκαν αναφέρονται και στα εισαγωγικά σχόλια. Ακόμη, μόνο η περίπτωση του χαρακτηριστικού mean περιέχει σχόλια πάνω στον κώδικα γιατί λειτουργεί σαν παράδειγμα. </br>
Μετά την δημιουργία των στηλών αφαιρέθηκε η στήλη "Magnitude" και όλες οι σειρές κάτω από αυτές που περιέχουν τα χαρακτηριστικά των 63 παραθύρων, ενώ ακόμη προστέθηκε μια νέα στήλη "Participant", η οποία περιέχει απλά την λέξη "Participant 1" στα κελιά της και θα μας βοηθήσει στην ταξινόμηση των δεδομένων στο έπειτα συγχωνευμένο αρχείο.

Υ.Γ. Ειδική είναι η περίπτωση του max-min, όπου απλά χρησιμοποιήθηκαν οι προηγούμενες στήλες για την δημιουργία της στήλης max-min.

```python
# Group the rows into chunks of 1000 rows and calculate the mean of the first column for each group
mean_values = par.groupby(par.index // 1000)['Magnitude'].mean().values

# Create a new DataFrame with the mean values for each group
mean_df = pd.DataFrame({'Mean': mean_values})

# Add the mean values to the original DataFrame by merging the two DataFrames based on the group index
par = pd.merge(par, mean_df, left_index=True, right_index=True, how='outer')
```

```python
# std (τυπική απόκλιση)
std_values = par.groupby(par.index // 1000)['Magnitude'].std().values
std_df = pd.DataFrame({'Std': std_values})
par = pd.merge(par, std_df, left_index=True, right_index=True, how='outer')

# skewness (ασυμμετρία κατανομής)
skew_values = par.groupby(par.index // 1000)['Magnitude'].apply(lambda x: skew(x)).values
skew_df = pd.DataFrame({'Skewness': skew_values})
par = pd.merge(par, skew_df, left_index=True, right_index=True, how='outer')

# max (μέγιστη τιμή)
max_values = par.groupby(par.index // 1000)['Magnitude'].max().values
max_df = pd.DataFrame({'Max': max_values})
par = pd.merge(par, max_df, left_index=True, right_index=True, how='outer')

# min (ελάχιστη τιμή)
min_values = par.groupby(par.index // 1000)['Magnitude'].min().values
min_df = pd.DataFrame({'Min': min_values})
par = pd.merge(par, min_df, left_index=True, right_index=True, how='outer')

# max-min (διαφορά μέγιστης και ελάχιστης τιμής)
par['Max-Min'] = par.iloc[:, 5] - par.iloc[:, 6]

# welch (εκτίμηση ισχύος φάσματος)
welch_values = par.groupby(par.index // 1000)['Magnitude'].apply(lambda x: welch(x)).values
welch_df = pd.DataFrame({'Welch': welch_values})
par = pd.merge(par, welch_df, left_index=True, right_index=True, how='outer')

# Περαιτέρω τακτοποίηση του αρχείου
par.drop(par.columns[0], axis=1,inplace=True)
par.drop(par.index[63:63000], axis=0,inplace=True)

# Εισαγωγή νέας στήλης ως ένδειξη προέλευσης αρχείου
data = 63*['Participant 1']
par.insert(0, 'Participant', data)

par
```

#### Η ειδική περίπτωση της εκτίμησης ισχύος φάσματος

Προφανώς η παραπάνω στήλη "Welch" δεν μπορεί να χρησιμοποιηθεί στην παρούσα της μορφή ως χαρακτηριστικό και έτσι πρέπει από αυτήν να εξάγουμε χρήσιμες πληροφοριές πάνω στα δεδομένα μας και αυτά να αποτέλεσουν τα χαρακτηριστικά μας. </br>

Πιο συγκεκριμένα, η στήλη "Welch" στην παρούσα της μορφή αποτελείται* από δύο πίνακες όπου ο ένας περιέχει συχνότητες και ο άλλος περιέχει τιμές PSD, δηλαδή Power Spectral Density ή Φασματική πυκνότητα ισχύος. Από αυτούς τους πίνακες, σαν ενδιάμεσο βήμα, θα δημιουργήσουμε δύο στήλες "Welch Frequency" και "Welch PSD", οι οποίοι θα περιέχουν τους πρώτους ή δεύτερους πίνακες της στήλης "Welch", αντίστοιχα. Έτσι, έπειτα με την βοήθεια αυτών των στηλών θα πάρουμε τις τιμές για τα νέα χαρακτηριστικά μας.

*Περισσότερες πληροφορίες για την συνάρτηση και το output της μπορείτε να βρείτε στο [documentation](https://docs.scipy.org/doc//scipy-1.3.3/reference/generated/scipy.signal.welch.html) της.

```python
# Function to extract arrays from the string
def extract_arrays(s):
    arrays = re.findall(r'(?<=\[)[^][]*(?=])', s)
    return [np.fromstring(a, sep=',') for a in arrays]

# Iterate over the 'Welch' column and extract the maximum values
array1 = []
array2 = []
for index, welch_data in par['Welch'].items():
    
    # Extract the arrays from the string
    first_array, second_array = extract_arrays(str(welch_data))

    # Append the arrays to their corresponding list
    array1.append(first_array)
    array2.append(second_array)
    
# Add the values to the DataFrame
par['Welch Frequency'] = array1
par['Welch PSD'] = array2
```

Σε αυτό το σημείο θα δημιουργήσουμε τις κατάλληλες συναρτήσεις για κάθε νέο χαρακτηριστικό ώστε χρησιμοποιώντας αυτές και τις στήλες "Welch Frequency" & "Welch PSD" να πάρουμε έπειτα τις νέες στήλες με τις τιμές των χαρακτηριστικών. Πιο συγκεκριμένα, θα χρειαστούμε συναρτήσεις για τον υπολογισμό των:
1. **Max PSD**: Μέγιστη τιμή του φάσματος ισχύος για κάθε παράθυρο.
2. **Max frequency**: Τιμή της συχνότητας που αντιστοιχεί στην μέγιστη τιμή φάσματος ισχύος και συχνά σχετίζεται με την κυρίαρχη συχνότητα ενός σήματος.
3. **Frequency centroid**: Μέτρο της σταθμισμένης μέσης συχνότητας του φάσματος ισχύος. </br>
   *> It represents the "center of mass" of the frequency distribution of the signal, weighted by the corresponding spectral power values. Mathematically, it is calculated as the sum of the product of frequency values and their corresponding power values divided by the sum of power values. The frequency centroid is a useful feature because it can provide information about the dominant frequency of the signal, and can help to differentiate signals with different frequency content.*
4. **Frequency bandwidth**: Μέτρο της έκτασης των συχνοτήτων στο φάσμα ισχύος. </br>
   *> It represents the width of the frequency distribution of the signal. Mathematically, it is calculated as the square root of the sum of the squared deviations of frequency values from the frequency centroid, weighted by the corresponding spectral power values, divided by the sum of power values. The frequency bandwidth is a useful feature because it can provide information about the variability or diversity of the frequency content of the signal.*
5. **Frequency skewness**: Μέτρο ασυμμετρίας του φάσματος ισχύος γύρω από την centroid frequency. </br>
   *> It represents the degree to which the frequency distribution of the signal is skewed to one side or the other of the centroid frequency. Mathematically, it is calculated as the sum of the cube of the deviations of frequency values from the frequency centroid, weighted by the corresponding spectral power values, divided by the product of the frequency bandwidth cubed and the sum of power values. The frequency skewness is a useful feature because it can provide information about the symmetry or asymmetry of the frequency content of the signal.*
6. **Spectral entropy**: Μέτρο της "ομαλότητας" του φάσματος ισχύος και του βαθμού διαταραχής ή τυχαιότητας ενός σήματος. </br>
   *> It represents the degree to which the spectral power values are evenly distributed across the frequency range. Mathematically, it is calculated as the negative sum of the product of each spectral power value and its logarithm (to base 2), normalized by the logarithm (to base 2) of the number of frequency bins. The spectral entropy is a useful feature because it can provide information about the degree of randomness or predictability in the frequency content of the signal.*

```python
# Define a function to compute the maximum psd (power spectral density)
def psd_max(freq,psd):
    max_index = np.argmax(psd)
    return psd[max_index]

# Define a function to compute the frequency corresponding to the maximum psd
def freq_max(freq,psd):
    max_index = np.argmax(psd)
    return freq[max_index]

# Define a function to compute the frequency centroid
def freq_centroid(freq, psd):
    return np.sum(freq * psd) / np.sum(psd)

# Define a function to compute the frequency bandwidth
def freq_bandwidth(freq, psd):
    centroid = freq_centroid(freq, psd)
    return np.sqrt(np.sum(((freq - centroid) ** 2) * psd) / np.sum(psd))

# Define a function to compute the frequency skewness
def freq_skewness(freq, psd):
    centroid = freq_centroid(freq, psd)
    bandwidth = freq_bandwidth(freq, psd)
    return np.sum(((freq - centroid) / bandwidth) ** 3 * psd) / np.sum(psd)

# Define a function to compute the spectral entropy
def spectral_entropy(psd):
    return -np.sum(psd * np.log2(psd))

# Compute the four features and add them as new columns in the dataframe
par['Max PSD'] = [psd_max(freq, psd) for freq, psd in zip(par['Welch Frequency'], par['Welch PSD'])]
par['Max Frequency'] = [freq_max(freq, psd) for freq, psd in zip(par['Welch Frequency'], par['Welch PSD'])]
par['Frequency Centroid'] = [freq_centroid(freq, psd) for freq, psd in zip(par['Welch Frequency'], par['Welch PSD'])]
par['Frequency Bandwidth'] = [freq_bandwidth(freq, psd) for freq, psd in zip(par['Welch Frequency'], par['Welch PSD'])]
par['Frequency Skewness'] = [freq_skewness(freq, psd) for freq, psd in zip(par['Welch Frequency'], par['Welch PSD'])]
par['Spectral Entropy'] = [spectral_entropy(psd) for psd in par['Welch PSD']]

# Remove Welch, Welch Frequency and Welch PSD columns
par.drop(["Welch", "Welch Frequency", "Welch PSD"], axis=1,inplace=True)
```
