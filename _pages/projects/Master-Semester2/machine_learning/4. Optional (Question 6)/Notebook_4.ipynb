## Εκπαίδευση δεδομένων

### Εισαγωγή δεδομένων εκπαίδευσης

Πριν ξεκινήσουμε οτιδήποτε θα πρέπει να εισάγουμε το αντίστοιχο "standardized_fix" αρχείο με τους Participants από right pocket:

### Εκπαίδευση με SVM

Θα χρησιμοποιήσουμε την μέθοδο SVM με τα βέλτιστα C=10000000 και γ=1e-06 στα αρχεία csv που περιέχουν την τροποποιημένη δραστηριότητα "stand or sit", κάνοντας εκπαίδευση με τα επεξεργασμένα από το "right pocket" και testing με τα επεξεργασμένα δεδομένα από τον "Avraam". </br>
Έτσι, έχουμε:

```python
# Rename training and testing dataframes
train_df = pars   # train with right pocket
test_df = par     # test with Avraam

# Extract features and labels from data frames
X_train = train_df.iloc[:, 2:14].values
y_train = train_df.iloc[:, 1].values
X_test = test_df.iloc[:, 2:14].values
y_test = test_df.iloc[:, 1].values

# Initialize classifier with optimal hyperparameters
svm = SVC(kernel='rbf', C=10000000, gamma=1e-06)

# Fit SVM classifier and predict labels
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)

# Calculate confusion matrix and accuracy
labels = np.unique(np.concatenate([y_train, y_test]))  # Get all unique labels from both training and testing data
cm = confusion_matrix(y_test, y_pred, labels=labels)

# Create DataFrame from confusion matrix and labels
cm_df = pd.DataFrame(cm, columns=labels, index=labels)

# Calculate overall accuracy and print confusion matrix
overall_acc = np.trace(cm) / np.sum(cm)
print(f"\nOverall accuracy: {overall_acc}")
print("\nConfusion matrix:\n")
print(cm_df)
```

### Σχόλια

Όπως είναι φανερό, επιτεύχθηκε πολύ χαμηλή ακρίβεια και υπάρχουν πολλές πιθανές αιτίες γι' αυτό:

- **Ποιότητα δεδομένων**: Τα νέα δεδομένα που συλλέξαμε μπορεί να περιέχουν περισσότερο θόρυβο και outliers σε σύγκριση με τα αρχικά δεδομένα. Αυτό μπορεί να καθιστά πιο δύσκολο για τον ταξινομητή SVM να γενικεύσει και να ταξινομήσει με ακρίβεια τα νέα δεδομένα.

- **Κατανομή δεδομένων**: Η κατανομή των νέων δεδομένων μπορεί να διαφέρει σημαντικά από τα αρχικά δεδομένα. Αν τα νέα δεδομένα περιέχουν διαφορετικά πρότυπα ή είναι μη ισορροπημένα ως προς την κατανομή των κλάσεων, ο ταξινομητής μπορεί να δυσκολευτεί να προσαρμοστεί σε αυτές τις αλλαγές και να επιτύχει ακριβείς προβλέψεις.

- **Overfitting**: Είναι πιθανό ότι ο ταξινομητής SVM υπερεκπαίδευσε (overfit) τα αρχικά δεδομένα, δηλαδή έμαθε υπερβολικά ακριβώς τα συγκεκριμένα πρότυπα και θόρυβο που υπάρχουν στο σύνολο εκπαίδευσης. Όταν εισάγονται νέα δεδομένα, ο ταξινομητής μπορεί να δυσκολευτεί να γενικεύσει αποτελεσματικά και η απόδοσή του μπορεί να υποχωρήσει. Αυτό συμβαίνει διότι ο ταξινομητής εστιάζει υπερβολικά στην αναπαράσταση των αρχικών δεδομένων, αντί να αντιλαμβάνεται τα γενικά πρότυπα που είναι ουσιώδη για την καλή απόδοση σε νέα δεδομένα.

- **Ελλειπής εκπαίδευση** : Αν τα νεότερα δεδομένα που συλλέξαμε είναι σημαντικά διαφορετικά από τα αρχικά δεδομένα, είναι πιθανό ότι ο ταξινομητής δεν έχει εκπαιδευτεί επαρκώς για να αντιμετωπίσει τα νέα πρότυπα ή έννοιες. Απαιτείται επιπλέον εκπαίδευση χρησιμοποιώντας τα νέα δεδομένα για να βελτιωθεί η απόδοση του ταξινομητή.
